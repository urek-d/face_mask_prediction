```python

from fastai.vision import *
import pandas as pd
import PIL
from IPython.display import Image as ShowImage
from matplotlib import pyplot as plt
```

```python
train_labels = pd.read_csv('train_labels.csv')
train_labels.head()
```

```python
ShowImage('images/rxgismgsvmaayzjarbfjaljhqmpbrt.jpg')
```

```python
#nombre d image a verifié
num_images_to_check=6
```

```python
#prendre des images au hasard
sample_df=train_labels.sample(n=num_images_to_check)
```

```python
for index,row in sample_df.iterrows():
    image_path='images/'+row['image']
    image_label = row['target']
    image = PIL.Image.open(image_path)
    plt.imshow(image)
    plt.title('Label:'+ str(image_label))
    plt.show()
```

```python
data = ImageDataBunch.from_df(path='images', df=train_labels,ds_tfms = get_transforms(),size=224, bs=32) 
data.show_batch(rows=2)
```

```python
learn = cnn_learner(data,models.resnet50,metrics=accuracy)
```

```python
lr_find(learn)
learn.recorder.plot(suggestion=True)
```

```python

learn.fit_one_cycle(3, 1e-4) 
```

```python
learn.unfreeze()
learn.fit_one_cycle(3, slice(1e-6, 1e-5))
```

```python
# lecture du fichier de soumission
ss = pd.read_csv('SampleSubmission.csv')
ss.head()
```

```python
# Create a test set to pass to our learner
test = (ImageList.from_df(path='', df=ss, folder='images'))
learn.data.add_test(test)
```

```python
# Get predictions
preds, labels = learn.get_preds(DatasetType.Test)
```

```python
# What do these look like?
preds[0] # The first prediction
```

```python
# we want p(Mask) for our submission, so we'll look at the second number in each prediction with p[1]
ss['target'] = [float(p[1]) for p in preds]
ss.head()
```

```python
ss.to_csv('submission.csv', index=False)
```
